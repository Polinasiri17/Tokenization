```python
import gradio as gr
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
def tokenize_text(text):
    tokens = tokenizer.tokenize(text)
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    return f"Tokens: {tokens}\n\nToken IDs: {token_ids}"
gr.Interface(
    fn=tokenize_text,
    inputs=gr.Textbox(label="Insert prompt here", lines=5, placeholder="Type a sentence..."),
    outputs=gr.Textbox(label="Tokenizer output will be here", lines=10),
    title="HF Transformer Demo"
).launch()
```
